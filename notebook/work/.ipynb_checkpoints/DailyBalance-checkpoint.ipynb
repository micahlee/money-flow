```ruby
require 'pg'

connection_name = ''
account_name = ''

classifier = nil
begin
    con = PG.connect(
      dbname: 'myapp_development',
      user: 'postgres',
      password: '',
      host: 'db'
      )
    
    rs = con.exec """
      WITH days AS (
        SELECT generate_series(
           (SELECT MIN(date) FROM transactions)::timestamp,
           CURRENT_TIMESTAMP,
           INTERVAL '1 day'
         )::date as day
      ),
      account_days AS (
        SELECT
          d.day,
          a.id as account_id,
          a.name as account_name
        FROM
          days d,
          accounts a
      ),
      daily_transactions AS (
        SELECT
          t.date,
          SUM(t.amount) as daily_total,
          a.id as account_id
        FROM
          transactions t
          JOIN accounts a on a.id = t.account_id
        WHERE
          t.pending = false
        GROUP BY
          t.date,
          a.id
      ),
      daily_balance AS (
        SELECT
          ad.day,
          ad.account_id,
          COALESCE(SUM(daily_total) OVER(
            PARTITION BY ad.account_id
            ORDER BY ad.day desc
          ), 0) as balance
        FROM
          account_days ad
          LEFT JOIN daily_transactions dt on dt.date::date = ad.day and dt.account_id = ad.account_id
          
      )
      SELECT
        db.day,
        c.name,
        a.name,
        (SELECT balance_current from accounts a2 where a2.id = db.account_id) - db.balance as balance
      FROM
        daily_balance db
        LEFT JOIN accounts a on a.id = db.account_id
        LEFT JOIN connections c on c.id = a.connection_id
      WHERE
        c.name = '#{connection_name}'
        AND a.name = '#{account_name}'
      ORDER BY
        db.day desc
      LIMIT 100
      """

    rs.each do |row|
      p row
    end

rescue PG::Error => e

    puts e.message 
    
ensure

    con.close if con
    
end
```
